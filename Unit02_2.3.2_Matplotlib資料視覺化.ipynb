# `Matplotlib` 資料視覺化

`matplotlib` 幾乎是標準 Python 畫圖套件了! 在有 `matplotlib` 之前, Python 要畫圖不那麼方便, 和 Python 很多套件一樣, 有許多方案, 但各家有不同的優缺點, 也沒有一套是大家都在用的。

而 `matplotlib` 仿 Matlab 式的畫圖方式, 讓很多人很快入手、並且功能相當完整。原作者是 John D. Hunter, 和很多 Python 的套件作者一樣, 他有博士學位。非常令人遺憾的是他在 2012 年因大腸直腸癌治療併發症過逝, 過逝時才 44 歲!


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1 畫個函數

前面說過 `plt.plot` 的基本用法是

    plt.plot(X,Y)

其中 X, Y 是分別是點 x 座標, y 座標形的的 list 或 array。

### 【重點】標準函數畫法

函數畫圖其實就是描點法, 而且再次用到陣列導向計算的概念。

我們先牛刀小試, 畫個 $\sin$ 的圖。

$$f(x) = \sin(x)$$

先來設個 x 的範圍, 就 -10 到 10, 給 100 個點。

```python
x = np.linspace(-10, 10, 100)
```

y 自然是 $\sin(x)$。

```python
y = np.sin(x)
```

畫圖!

```python
fig,ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

### 【練習】函數畫圖

這次我們來畫:

$$f(x) = \frac{\sin(5x)}{1+x^2}$$

```python
x = np.linspace(0, 10, 200)
y = np.sin(5*x) / (1+x**2)
fig,ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## 4 如果畫個參數式圖形

其實參數式圖形基本上也是一堆 x, y 的座標, 只是 x(t), y(t) 都是 t 的函數, 我們可以想成 t 是時間。

假設我們的

$$ -2\pi \leq t \leq 2\pi$$

要畫個半徑是 r 的圓, x(t), y(t) 就是這樣:

* $x(t) = r\cos(t)$
* $y(t) = r\sin(t)$

我們來試試事情是不是如我們想的這樣...

```python
r = 3
t = np.linspace(-2*np.pi, 2*np.pi, 200)
x = r*np.cos(t)
y = r*np.sin(t)
```

```python
fig,ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

真的成功了... 等等, `matplotlib` 怎麼畫成橢圓呢? 我們來調整一下。

```python
fig,ax = plt.subplots()
ax.plot(x, y)
ax.set_aspect("equal")
plt.show()
```

再來我們可以想一下, 如果 r 不是故定的呢? 也就是它會變長變短, 是不是會畫出很多特別的圖呢? 其實很多特別的曲線不過就是找個試合的 r(t), 我們來隨意試試看。

```python
r = np.sin(5*t)
t = np.linspace(-2*np.pi, 2*np.pi, 200)
x = r*np.cos(t)
y = r*np.sin(t)
fig,ax = plt.subplots()
ax.plot(x, y)
ax.set_aspect("equal")
plt.show()
```

是不是很有意思呢? 你也可以自己試試不同的變化。

## 5 `subplot` 畫多個圖

我們每次畫圖的時候, `matplotlib` 就弄 1 個 figure 畫圖區出來, 裡面可以有很多子圖, 在 figure 裡叫 axes。目前我們都只有 1 個 figure 內含 1 張圖, 所以都不用設, 現在我想畫 4 張圖時。我們就要先想好「陣式」。

比如說 2x2 這樣排列的 4 張圖。

```python
x = np.linspace(0, 10, 100)

fig,ax = plt.subplots(2, 2)
ax[0,0].plot(x, np.sin(x), "r")
ax[0,1].plot(x, np.sin(3*x), "g")
ax[1,0].scatter(x, np.random.rand(100), color="y", alpha=0.5)
ax[1,1].bar(range(10), np.random.randint(1, 30, 10), color="b")
plt.show()
```

## 7 Marker*

所謂 'marker' 就是原來的點, 我們特別標示出來, 為了不要讓圖形太複雜, 我們新換個點比較少的例子。

```python
x = range(20)
y = np.random.randn(20)
fig,ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x, y, marker="o")
plt.show()
```

#### marker 可以設的參數

| 參數 | 說明 |
|---|---|
| `marker` | marker 的風格 |
| `markeredgecolor` (`mec`) | 邊線顏色 |
| `markeredgewidth` (`mew`) | 邊線寬度 |
| `markerfacecolor` (`mfc`) | marker 的顏色 |
| `markerfacecoloralt` (`mfcalt`) | marker 替換色 |
| `markersize` (`ms`) | marker 大小 |
| `markevery` | 隔多少畫一個 marker |

```python
fig,ax = plt.subplots()
ax.plot(x, y, lw=3, marker="o", mfc="y", mec="k", ms=10)
plt.show()
```

比較奇特的參數是 `markevery`, 這是說我們每多少筆資料要畫一個 marker。為了示範, 我們拿之前的例子來看看。

```python
x = np.linspace(0, 10, 200)
y = np.sin(5*x) / (1+x**2)
fig,ax = plt.subplots()
ax.plot(x,y,marker="o")
plt.show()
```

後面的完完全全連在一起了, 我們現在隔 10 個畫一個 marker 試試。

```python
fig,ax = plt.subplots()
ax.plot(x,y, marker="o", markevery=8)
plt.show()
```

## 9 畫圖區的設定*

#### 說明文字的標示

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.sin(x)
fig,ax = plt.subplots()
ax.plot(x, y, lw=5)
ax.set_title("A simple line plot for $sin(x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$sin(x)$")
plt.show()
```

#### 邊界的設定

```python
fig,ax = plt.subplots()
ax.plot(x,y,lw=5)
ax.set_title("A simple line plot for $sin(x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$sin(x)$")
ax.set_xlim(-6, 6)
ax.set_ylim(-1, 1)
plt.show()
```

#### xticks

```python
fig,ax = plt.subplots()
ax.plot(x, y, lw=5)
ax.set_title("A simple line plot for $sin(x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$sin(x)$")
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(["$-2\pi$", "$-\pi$", 0, "$\pi$", "$2\pi$"])
plt.show()
```

## 10 取得現在工作中 axes*

我們有時要設 axes 的背景啦等等的資訊。這時就要取得現在工作中的 axes。這一般有兩種方式, 第一種是設 `subplot` 時可以取得:

    fig, ax = plt.subplot()
    
另一種是用 `gca` 函數:

    ax = plt.gca()

```python
ax = plt.gca()
ax.set_facecolor('#69b8bb')
ax.set_xlim(-6, 6)
ax.set_ylim(-1.2, 1.2)
plt.plot(x,y, lw=5, c='white')
plt.show()
```

#### 移動 x, y 座標軸

```python
ax = plt.gca()
ax.set_facecolor('#69b8bb')
ax.set_xlim(-6, 6)
ax.set_ylim(-1.2, 1.2)
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
plt.plot(x, y, lw=5, c='white')
plt.show()
```
